```python
import seaborn as sns 
```

```python
%matplotlib inline
```

```python
#distplot jointplot pairplot rugplot kdeplot
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
#tips.info()
```

```python
sns.distplot(tips['total_bill'],bins=20,rug=True,hist=False)
```

```python
sns.distplot(tips['total_bill'],bins=10,kde=False,hist=True)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='size',data=tips,kind='hex',color='blue',ylim=4)
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='scatter',color='black',xlim=4)
```

```python
sns.pairplot(tips,hue='sex',palette='rocket_r',kind='scatter',dropna=True)
```

```python
sns.rugplot(tips['tip'],color='red',height=5)
sns.rugplot(tips['total_bill'],color='yellow',height=3)
```

```python
#categorical
#factorplot boxplot violinplot stripplot swarmplot barplot countplot
```

```python
sns.barplot(x='tip',y='sex',data=tips,hue='day')
```

```python
import numpy as np 
sns.barplot(x='tip',y='sex',data=tips,hue='time',estimator=np.min)
```

```python
sns.countplot(y='day',data=tips,hue='smoker',dodge=True)
```

```python
sns.countplot(x='smoker',data=tips,hue='day')
```

```python
sns.boxplot(x='tip',y='day',data=tips,hue='smoker',color='green',linewidth=1,width=0.5,fliersize=7)
```

```python
sns.violinplot(x='size',y='day',data=tips,hue='time',bw='scott',split=True,dodge=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='smoker',jitter=False,size=8,linewidth=2,dodge=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='smoker',dodge=True)
```

```python
sns.swarmplot(x='time',y='tip',data=tips,hue='day',dodge=True)
```

```python
import matplotlib.pyplot as plt 
```

```python
iriss=sns.load_dataset('iris')
```

```python
iriss.head()
```

```python
pg=sns.PairGrid(iriss)
pg.map_diag(plt.hist)
pg.map_upper(plt.scatter)
pg.map_lower(sns.kdeplot)
```

```python
tips.head()
```

```python
g=sns.FacetGrid(tips,row='smoker',col='time',hue='day')
g.map(plt.scatter,"tip","total_bill").add_legend()
```

```python
j=sns.JointGrid(data=tips,x='tip',y='total_bill')
j.plot(sns.kdeplot,sns.distplot)
```
